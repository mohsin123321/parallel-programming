```python
import numpy as np
from matplotlib import pyplot as plt

center_1 = np.array([1,1])
center_2 = np.array([3000,3000])
center_3 = np.array([5000,1])


sd = 400
data_1 = sd * np.random.randn(30000, 2) + center_1
data_2 = sd * np.random.randn(30000, 2) + center_2
data_3 = sd * np.random.randn(40000, 2) + center_3

data = np.concatenate((data_1, data_2, data_3), axis = 0)


plt.scatter(data[:,0], data[:,1], s=7)
```

```python
print(len(data))
```
